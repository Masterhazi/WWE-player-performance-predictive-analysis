```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import time
from tqdm import tqdm
from google import genai
from google.genai import types

# Initialize DataFrame and Google GenAI Client
df = pd.DataFrame(columns=['Event Name', 'Winner', 'Loser', 'Title Match'])

genai_client = genai.Client(api_key='AIzaSyCzAKfRu39JW1UwD5i-8F0mPbp-850YSF8')

event_links

# Step 2: Loop through links and scrape data
for link in tqdm(event_links, desc="Scraping events", ncols=100):
    try:
        # Scrape HTML of the event page
        response = requests.get(link, headers=headers)
        if response.status_code != 200:
            print(f"Failed to fetch data from {link}")
            continue
        
        html_data = response.text[:]  # Limit to 10,000 characters to avoid overflow
        
        # Pass HTML to Gemini for extraction
        prompt = f"""
        **Extract the following details from the HTML given if you did not receive any html code then dont do anything instead respond saying "HTML data not received please check the flow properly":**  
        1. Event Name – Found in `<h2 class="section-title">`, e.g., "WWE RAW: Day 1 (2024)" simplify the event name to WWE RAW, this would be enough.  
        2. Winner – Found in `<div class="winning-side winner">` under `<span class="wrestler-name">`.  
        3. Loser – Found in `<div class="losing-side">` under `<span class="wrestler-name">`.  
        4. Title Match – Presence of `<span class="champion">C</span>` indicates a title match.  

        **Special Cases:**  
        - Tag Team Matches or matches with multiple players in one team:  
        - If multiple players are involved (e.g., Roman Reigns & Solo Sikoa vs Jimmy Uso & Jey Uso):  
            - List each winner and loser separately. I dont want the players to be listed like roman reigns, solo sikoa instead i need roman reigns in separate row and solo sikoa in separate row  
            - Example: If Roman Reigns & Solo Sikoa win, output should be:  
            - in a new row Winner = Roman Reigns, Loser = Jimmy Uso  
            - in a new row Winner = Solo Sikoa, Loser = Jey Uso  
        -If it is a handicap match 
            -Example: Roman Reigns vs Jey uso and jimmy uso
            if roman reign won then 
            winner = roman reigns
            loser  = any one of them
            if jey uso and jimmy uso win
            winner = any one of them
            loser = Roman reigns


        **Output Format:**  
        Event Name | Winner | Loser | Title Match (Yes/No)  

        **HTML:**  
        {html_data} 
        """

        contents = [
            types.Content(
                role="user",
                parts=[types.Part(text=prompt)]
            )
        ]

        config = types.GenerateContentConfig(
            temperature=1,
            max_output_tokens=8192,
            response_mime_type="text/plain"
        )

        response = genai_client.models.generate_content_stream(
            model="gemini-2.0-flash-lite",
            contents=contents,
            config=config
        )

        result = "".join(chunk.text for chunk in response)
        print(f'''This is \n{result}''')
        
        # Parse LLM Output
        for line in result.split("\n"):
            if line.strip():
                parts = line.split(" | ")
                if len(parts) == 4:
                    df.loc[len(df)] = parts

        print(f"✅ Processed event from {link}")
        print(f'''This is DataFrame \n{df}''')

    except Exception as e:
        print(f"❌ Error processing {link}: {e}\n {result}")

    # Delay to avoid rate limiting
    time.sleep(2)

# Step 3: Save to CSV once at the end
if os.path.exists('wwe_matches2020.csv'):
    df.to_csv('wwe_matches2020.csv', mode='a', header=False, index=False)
else:
    df.to_csv('wwe_matches2020.csv', index=False)

print("✅ All data appended to 'wwe_matches2020.csv'")
print('🕶 Bravo 6 going dark 🕶')
```
